# 🌱 Part 0 — Foundations

## 0.1 What is DNA? 🧬

**DNA (Deoxyribonucleic Acid)** is the molecule that carries the genetic instructions for the development, functioning, growth, and reproduction of all known living organisms.

### The Alphabet of Life
DNA uses only **4 nucleotide bases** — think of them as a 4-letter alphabet:

| Base | Symbol | Pairs with |
|------|--------|------------|
| Adenine | **A** | T |
| Thymine | **T** | A |
| Cytosine | **C** | G |
| Guanine | **G** | C |

A DNA **sequence** is just a long string of these four letters, e.g.:
```
ATGCCCCAACTAAATACCGCCGTATGACCC...
```

### From DNA to Gene to Protein
```
DNA  ──(transcription)──►  mRNA  ──(translation)──►  Protein  ──►  Trait / Disease
```

### Why does the sequence matter for AI?
Different **species** and different **gene families** have distinctive patterns in their DNA sequences. By learning these patterns, an AI model can:
- Classify a DNA fragment to its species (Human / Dog / Chimpanzee)
- Identify the gene family it belongs to
- Predict functional properties (e.g., protein-coding vs. non-coding)

## 0.2 What is the Phenome? 📊

| Term | Definition |
|------|------------|
| **Genome** | All of an organism's DNA (~3 billion base pairs in humans) |
| **Transcriptome** | All RNA molecules expressed from the genome |
| **Proteome** | All proteins produced in a cell/organism |
| **Phenome** | The **complete set of observable characteristics** (phenotypes) of an organism: height, blood pressure, disease status, lab values, etc. |

### Phenome in Healthcare
In clinical AI, the phenome includes:
- **Vital signs** — heart rate, blood pressure, temperature
- **Lab values** — cholesterol, blood sugar, creatinine
- **Imaging features** — from X-ray, MRI, ECG
- **Clinical outcomes** — diagnosis, hospitalization, mortality

> 💡 **Key Insight:** The heart attack dataset we'll use is a perfect example of *phenomic data* — structured measurements of a patient's observable physiological state.

## 0.4 The ML Pipeline — Big Picture 🗺️

Every problem in this hackathon follows the same fundamental pipeline:

$$\underbrace{\text{Raw Data}}_{\text{Input}} \xrightarrow{\text{EDA}} \underbrace{\text{Features } \mathbf{X}}_{\text{n\_samples × n\_features}} \xrightarrow{\text{Model}} \hat{y} \xrightarrow{\text{Evaluate}} \underbrace{\text{Score}}_{\text{AUC, F1, Acc}}$$

### Key Mathematical Notation
- $\mathbf{X} \in \mathbb{R}^{n \times d}$ — feature matrix ($n$ samples, $d$ features)
- $\mathbf{y} \in \{0, 1, ..., K\}^n$ — target labels
- $f: \mathbf{X} \rightarrow \hat{\mathbf{y}}$ — the model we want to learn
- $L_{\text{loss}}(\mathbf{y}, \hat{\mathbf{y}})$ — loss function we minimize during training

## 1.2 Feature Engineering: k-mer Frequency 🔬

### What is a k-mer?

A **k-mer** is any contiguous substring of length $k$ from a DNA sequence.

**Example** with $k = 3$ ("codons"):
```
Sequence:  A T G C C C C A A C T
3-mers:    ATG TGC GCC CCC CCA CAA AAC ACT
```

### The Feature Vector

For alphabet $\{A, T, C, G\}$ and length $k$, there are $4^k$ possible k-mers.

$$\mathbf{x}_i = \left[ \frac{\text{count}(kmer_1)}{N_i},\ \frac{\text{count}(kmer_2)}{N_i},\ \ldots,\ \frac{\text{count}(kmer_{4^k})}{N_i} \right]$$

where $N_i$ is the total number of k-mers in sequence $i$.

| k | Features ($4^k$) | Notes |
|---|-----------------|-------|
| 1 | 4 | Just base frequencies |
| 2 | 16 | Dinucleotide patterns |
| **3** | **64** | **Codon-level (our choice)** |
| 4 | 256 | Higher-order patterns |
| 6 | 4096 | Very high-dimensional |

> We use **k=6** via sklearn's `CountVectorizer` on k-mer strings — a simple but powerful NLP-style approach!


````python
# ─── k-mer Feature Extraction ─────────────────────────────────────────────────
from sklearn.feature_extraction.text import CountVectorizer

def get_kmers(sequence, k=6):
    """
    Converts a DNA sequence into a space-separated string of k-mers.
    This lets us use sklearn's CountVectorizer as a k-mer counter!

    Args:
        sequence: DNA string (e.g., 'ATGCCC...')
        k: k-mer length (default 6)
    Returns:
        Space-separated k-mer string (e.g., 'atgccc tgccca gcccat ...')
    """
    seq = sequence.lower()  # normalize to lowercase
    return ' '.join([seq[i:i+k] for i in range(len(seq) - k + 1)])

# Test it
example_seq = 'ATGCCCCAACTAAATACCGCCGTATGACCC'
print('Example sequence:', # Fill in your code here)
print('k=3 mers:', # Fill in your code here)
print('k=6 mers (first 5):', ' '.join(get_kmers(example_seq, k=6).split()[:5]), '...')
````

## 1.3 Dimensionality Reduction — PCA & UMAP 🌐

### Why Reduce Dimensions?
With $4^6 = 4096$ possible 6-mers, our feature space is high-dimensional. Dimensionality reduction helps us:
1. **Visualize** data in 2D/3D
2. **Remove noise** and redundancy
3. **Speed up** downstream models

### PCA — Principal Component Analysis

PCA finds orthogonal directions of **maximum variance** in the data.

$$\mathbf{Z} = \mathbf{X} \mathbf{W}$$

where $\mathbf{W} \in \mathbb{R}^{d \times k}$ are the top-$k$ **eigenvectors** of the covariance matrix $\mathbf{\Sigma} = \frac{1}{n}\mathbf{X}^T\mathbf{X}$.

The **explained variance ratio** tells us how much information each PC captures:
$$\text{EVR}_i = \frac{\lambda_i}{\sum_j \lambda_j}$$

where $\lambda_i$ are the eigenvalues.

### UMAP — Uniform Manifold Approximation and Projection

## 1.4 Feature Selection — LASSO (L1 Regularization) ✂️

### The Problem of High Dimensions
With thousands of k-mer features, many are **redundant or irrelevant**. Too many features causes:
- Overfitting
- Slow training
- Hard to interpret results

### LASSO — Least Absolute Shrinkage and Selection Operator

LASSO adds an **L1 penalty** to the loss function:

$$
L_{\text{LASSO}}(\beta)
=
\frac{1}{n}\sum_{i=1}^n (y_i-\hat y_i)^2
+
\lambda\sum_{j=1}^d |\beta_j|
$$

The L1 penalty forces many $\beta_j$ coefficients to **exactly zero**, effectively selecting a sparse set of features.

| Regularization | Penalty | Effect |
|---|---|---|
| **LASSO (L1)** | $\lambda \|\beta\|_1$ | Sets many weights to 0 → feature selection |
| **Ridge (L2)** | $\lambda \|\beta\|_2^2$ | Shrinks weights but keeps all |
| **Elastic Net** | $\alpha \|\beta\|_1 + (1-\alpha)\|\beta\|_2^2$ | Best of both worlds |

### The Regularization Path
As $\lambda$ increases → more features are zeroed out → simpler model.

## 1.5 Tabular ML — XGBoost (Main Model) 🏆

### Why XGBoost Wins Kaggle Competitions

**XGBoost** (Extreme Gradient Boosting) is an ensemble of **decision trees** trained sequentially. Each tree corrects the errors of the previous one.

### The Math Behind XGBoost

The objective function to minimize at round $t$:

$$
L^{(t)}
=
\sum_{i=1}^{n}
\ell\!\left(
y_i,\,
\hat{y}_i^{(t-1)} + f_t(\mathbf{x}_i)
\right)
+
\Omega(f_t)
$$


where:
- $\ell$ is the loss (e.g., cross-entropy for classification)
- $f_t$ is the $t$-th tree to be learned
- $\Omega(f_t) = \gamma T + \frac{1}{2}\lambda \|\mathbf{w}\|^2$ is the regularization term ($T$ = number of leaves)

Using second-order Taylor expansion, the optimal leaf weight becomes:

$$w_j^* = -\frac{\sum_{i \in I_j} g_i}{\sum_{i \in I_j} h_i + \lambda}$$

where $g_i = \partial_{\hat{y}} \ell$ (gradient) and $h_i = \partial_{\hat{y}}^2 \ell$ (hessian).

### Key Hyperparameters
| Parameter | Meaning | Tuning Tip |
|---|---|---|
| `n_estimators` | Number of trees | Use early stopping |
| `max_depth` | Max tree depth | 3–8, higher = more complex |
| `learning_rate` | Shrinkage per step | Smaller = more trees needed |
| `subsample` | Row subsampling | 0.6–0.9, prevents overfitting |
| `colsample_bytree` | Feature subsampling | 0.6–0.9 |
| `min_child_weight` | Min samples in leaf | Higher = more conservative |

## 1.7 Interpretability — SHAP Values 🔎

### Why Interpretability Matters in Healthcare?
A black-box model that says "60% chance of disease" is not enough. Clinicians need to know **why**.

### SHAP — SHapley Additive exPlanations

SHAP is rooted in **game theory**. The Shapley value for feature $j$ is:

$$\phi_j = \sum_{S \subseteq F \setminus \{j\}} \frac{|S|!(|F|-|S|-1)!}{|F|!} \left[ f(S \cup \{j\}) - f(S) \right]$$

In plain English: **how much does feature $j$ contribute to the model output, averaged over all possible feature subsets?**

### SHAP Properties
| Property | Meaning |
|---|---|
| **Local accuracy** | SHAP values sum to the model's prediction |
| **Missingness** | Zero SHAP if feature has no effect |
| **Consistency** | More important features always get higher SHAP |

## 2.2 Handling Class Imbalance 🎯

### Why Class Imbalance Is a Problem

If 95% of patients don't have heart disease, a model that always says "No" gets 95% accuracy — but it's **completely useless** clinically!

In our dataset, let's check the balance:

### Strategies for Imbalanced Classification

| Method | How it Works | Pros | Cons |
|--------|-------------|------|------|
| **Class weights** | Penalize misclassifying minority class more | No data modification | Slight accuracy drop |
| **SMOTE** | Synthesize new minority samples | More training data | May create noise |
| **Threshold tuning** | Adjust decision boundary | Simple post-hoc | Needs calibrated probs |
| **Focal Loss** | Down-weight easy examples | Native in PyTorch | Requires custom loss |

### Class-Weighted Loss (Main Method)

For class $c$, the weight is:

$$w_c = \frac{n_{\text{total}}}{n_{\text{classes}} \times n_c}$$

where $n_c$ is the number of samples in class $c$. This ensures **each class contributes equally** to the loss regardless of size.

### SMOTE — Synthetic Minority Over-sampling Technique

SMOTE creates synthetic samples by interpolating between nearest neighbors:

$$\mathbf{x}_{\text{new}} = \mathbf{x}_i + \lambda \cdot (\mathbf{x}_{nn} - \mathbf{x}_i), \quad \lambda \sim U(0, 1)$$

where $\mathbf{x}_{nn}$ is a random nearest neighbor of the minority sample $\mathbf{x}_i$.

## 2.3 Feature Selection — LASSO & Elastic Net 🎛️

### Elastic Net

Elastic Net combines L1 and L2 penalties:

$$L_{\text{EN}}(\beta)
=
\frac{1}{2n}\|\mathbf{y} - \mathbf{X}\beta\|_2^2
+
\lambda \left[
\alpha \|\beta\|_1
+
\frac{1-\alpha}{2}\|\beta\|_2^2
\right]$$

where:
- $\alpha = 1$ → pure LASSO
- $\alpha = 0$ → pure Ridge
- $0 < \alpha < 1$ → Elastic Net

**Why Elastic Net?** When features are correlated (e.g., age and max heart rate), LASSO picks one arbitrarily. Elastic Net keeps groups of correlated features together.

### Early Stopping — A Critical Trick

Instead of manually choosing the number of trees, **early stopping** automatically stops training when the validation metric stops improving.

```
Validation AUC
     ↑
     │    ╭──╮
     │   ╱    ╲_____  ← overfitting starts here
     │  ╱
     │ ╱
     │╱
     └──────────────────→ Number of Trees
        ↑ STOP HERE (best_iteration)
```

### Evaluation Metrics for Binary Classification

| Metric | Formula | When to use |
|--------|---------|-------------|
| **Accuracy** | $\frac{TP+TN}{N}$ | Balanced classes |
| **Precision** | $\frac{TP}{TP+FP}$ | When false positives are costly |
| **Recall (Sensitivity)** | $\frac{TP}{TP+FN}$ | When false negatives are costly |
| **F1-Score** | $\frac{2 \cdot P \cdot R}{P + R}$ | Imbalanced classes |
| **AUC-ROC** | Area under ROC curve | Overall discrimination ability |

In medicine, **Recall** is often most critical — missing a heart attack (FN) is far worse than a false alarm (FP)!

## 2.7 Survival Analysis Primer — Cox Proportional Hazards 📈

### Why Survival Analysis?
Classification tells us *if* a patient will have a heart attack. **Survival analysis tells us *when*.**

Clinical questions survival analysis answers:
- What is the 5-year survival probability for a patient with these characteristics?
- Which risk factors accelerate disease progression?
- How does treatment affect the time to event?

### Cox Proportional Hazards Model

The hazard function (instantaneous risk of event at time $t$):

$$h(t | \mathbf{x}) = h_0(t) \cdot \exp\left(\sum_{j=1}^p \beta_j x_j\right)$$

where:
- $h_0(t)$ = **baseline hazard** (non-parametric, unspecified)
- $\exp(\beta_j)$ = **hazard ratio** for feature $j$

### Hazard Ratio Interpretation
$$HR_j = e^{\beta_j}$$

| HR | Meaning |
|----|--------|
| HR > 1 | Feature increases risk (e.g., HR=2 → 2× more likely to have event) |
| HR = 1 | No effect |
| HR < 1 | Feature is protective |

> **Note:** The heart dataset doesn't include time-to-event, so we'll simulate a realistic survival analysis using the target variable as the event indicator and age as a proxy for follow-up time. In a real clinical study, you would have actual survival times.


````python
# ─── Cox PH Analysis (using age as proxy time, output as event) ───────────────
# In a real study: T = days/months until heart attack; E = 1 if event occurred
# Here we demonstrate the method using age as surrogate time

cox_df = heart[['age', 'target', 'sex', 'cp', 'chol', 'thalach', 'oldpeak', 'ca', 'thal']].copy()
cox_df.columns = [
    'duration', 'event', 'sex', 'chest_pain', 'cholesterol',
    'max_hr', 'st_depression', 'vessels', 'thalassemia'
]

# Fit Cox PH model
cph = CoxPHFitter(penalizer=0.1)
cph.fit(cox_df, duration_col='duration', event_col='event')

print('═══ Cox Proportional Hazards Model Summary ═══')
cph.print_summary()

# Plot hazard ratios
fig, ax = plt.subplots(figsize=(10, 6))
cph.plot(ax=ax)
ax.set_title('Cox PH Hazard Ratios with 95% Confidence Intervals\n(dots right of 1.0 = risk factors; left = protective)',
              fontweight='bold')
ax.axvline(x=1, color='red', linestyle='--', linewidth=1.5, label='HR=1 (no effect)')
ax.legend()
plt.tight_layout()
plt.show()
````
